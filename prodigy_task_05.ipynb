```python
# IMPORT NECESSARY LIBRARIES 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# remove unnecessary columns
df = df[['Severity', 'Start_Time', 'End_Time', 'Start_Lat', 'Start_Lng', 'Weather_Condition', 'Visibility(mi)', 'Precipitation(in)']]
df.head()
```

```python
# Convert time columns to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce', format='mixed')
df['End_Time'] = pd.to_datetime(df['End_Time'], errors='coerce', format='mixed')
```

```python
# Extract time-based features
df['Hour'] = df['Start_Time'].dt.hour
df['Day'] = df['Start_Time'].dt.day
df['Month'] = df['Start_Time'].dt.month
df['Weekday'] = df['Start_Time'].dt.weekday
```

```python
# EDA: Plot accidents by time of day
plt.figure(figsize=(10, 5))
sns.histplot(df['Hour'], bins=24, kde=True, color='blue')
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Time of Day")
plt.show()
```

```python
# Accidents by weather condition
plt.figure(figsize=(12, 5))
weather_counts = df['Weather_Condition'].value_counts().head(10)
sns.barplot(x=weather_counts.index, y=weather_counts.values, palette='coolwarm')
plt.xticks(rotation=45)
plt.xlabel("Weather Condition")
plt.ylabel("Accident Count")
plt.title("Top 10 Weather Conditions Leading to Accidents")
plt.show()
```

```python
import folium
from folium.plugins import HeatMap
```

```python
# Heatmap of accident locations
map = folium.Map(location=[df['Start_Lat'].mean(), df['Start_Lng'].mean()], zoom_start=5)
heat_data = list(zip(df['Start_Lat'], df['Start_Lng']))
HeatMap(heat_data[:10000]).add_to(map)  # Using first 10,000 points for performance
map.save("accident_hotspots.html")
print("Heatmap saved as accident_hotspots.html")
```

```python
# Correlation matrix for weather factors
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Severity', 'Visibility(mi)', 'Precipitation(in)']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
